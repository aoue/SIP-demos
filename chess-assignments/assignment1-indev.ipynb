```python
def DFS(b:chess.Board):
    frontier = [b]
    visited = [chess.A1]
    while frontier:
        current = frontier.pop()
        if current.move_stack:
            visited.append(current.move_stack[-1].to_square)
        if goal(current):
            print(f"SUCCESS frontier={frontier}")
            print(f"SUCCESS visited={visited}")
            return visited
        for m in current.legal_moves:
            if m.to_square in visited:
                continue
            n = current.copy()
            n.push(m)
            n.turn = chess.WHITE
            frontier.append(n)
    print(f"FAIL visited={frontier}")
    return None
```

```python
# Solution

# Right now this returns a Board which represents the final state
# Have them return a list of all visited nodes?

"""
change BFS algorithm so that nodes are not visited more than once

we get duplicates in visited because many moves have the same origin, even if they result in different board positions
"""

def BFS(b:chess.Board):
    frontier = [b]
    visited = [chess.A1]

    visited_set = set()

    while frontier:
        # take the oldest board situation from the frontier
        current = frontier.pop(0) # The only difference
        # print(current.move_stack)

        # if a move has already been made on this board (move stack is essentially a history of moves)
        if current.move_stack:
            # append to visited: the int id of the square that was just moved from
            visited.append(current.move_stack[-1].to_square)
            # visited_set.add(current.move_stack[-1].to_square)

        # check if we've met the goal condition
        if goal(current):
            return visited
        
        # goes through every legal move 'm' from this board position
        for m in current.legal_moves:
            # check whether the destination of 'm' has already been moved to 
            if m.to_square in visited:
                # if it has already been moved to, then skip the move
                continue

            # if it hasn't already been moved to

            # create a new board copy and update it after 'm'
            n = current.copy()
            n.push(m)
            n.turn = chess.WHITE

            # append this new situation to the frontier
            frontier.append(n)

    print(f"FAIL visited={frontier}")
    return None
```

```python
# Board2 class which keeps track of captured pieces and points
# Instead of making a subclass, may be better to use an internal Board
# This would prevent the use of superclass functions that don't know about the points 
class Board2(chess.Board):
    # Not standard, but this is for testing purposes
    point_map = {
        chess.PAWN : 1,
        chess.ROOK : 2,
        chess.KNIGHT : 3,
        chess.BISHOP : 4,
        chess.QUEEN : 5,
        chess.KING : 6,
    }
    def __init__(self, fen=None):
        super().__init__(fen)
        self._points = 0
    
    @property
    def points(self):
        return self._points
    
    def push(self, move:chess.Move):
        p = self.piece_at(move.to_square)
        super().push(move)
        if p:
            self._points += self.point_map[p.piece_type]

    def copy(self):
        new = super().copy()
        new._points = self._points
        return new

# Placeholder
def heuristic(b):
    return 0

# Placeholder UCS since it doesn't need a heuristic
def UCS(b: Board2):
    frontier = [(b, 0)]
    visited = [chess.A1]
    while frontier:
        current = frontier.pop(0)[0]
        if current.move_stack:
            visited.append(current.move_stack[-1].to_square)
        if goal(current):
            return current
        for m in current.legal_moves:
            if m.to_square in visited:
                continue
            n = current.copy()
            n.push(m)
            n.turn = chess.WHITE

            next_piece = current.piece_at(m.to_square)
            cost = Board2.point_map[next_piece.piece_type] if next_piece else 0
            cost += current.points
            # To turn this into A*, uncomment the following
            # cost += heuristic(n) - heuristic(current)

            for i, (old_board, old_cost) in enumerate(frontier):
                if old_board.move_stack[-1].to_square == m.to_square:
                    if cost < old_cost:
                        frontier.pop(i)
                        frontier.append((n, cost))
                    break
            else:
                frontier.append((n, cost))
            # There are better ways to do this, but this works for now
            frontier.sort(key=lambda x: x[1])
    return None
```
